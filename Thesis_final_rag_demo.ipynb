## 2. Partition PDF and store to Document

```python
import os
import requests
import zipfile
import io

# Define the URLs for the required NLTK data files
nltk_data_urls = {
    'punkt': 'https://raw.githubusercontent.com/nltk/nltk_data/gh-pages/packages/tokenizers/punkt.zip',
    'stopwords': 'https://raw.githubusercontent.com/nltk/nltk_data/gh-pages/packages/corpora/stopwords.zip'
}


# Define the data directory
nltk_data_dir = os.path.expanduser('/content/nltk_data')

# Ensure the directory exists
if not os.path.exists(nltk_data_dir):
    os.makedirs(nltk_data_dir)

for data_name, url in nltk_data_urls.items():
    # Create the appropriate subdirectory in the NLTK data directory
    subdir = os.path.join(nltk_data_dir, data_name)
    if not os.path.exists(subdir):
        os.makedirs(subdir)

    # Download the data file
    print(f"Downloading {data_name} from {url}...")
    response = requests.get(url)
    response.raise_for_status()  # Ensure the request was successful

    # Extract the zip file into the appropriate subdirectory
    with zipfile.ZipFile(io.BytesIO(response.content)) as z:
        z.extractall(subdir)
    print(f"{data_name} has been downloaded and extracted to {subdir}.")

# Set the NLTK data path
import nltk
nltk.data.path.append(nltk_data_dir)
```

```python
from unstructured.partition.pdf import partition_pdf
from langchain_core.documents import Document
from langchain_text_splitters import CharacterTextSplitter
import nltk
# Set the NLTK data path
#nltk.data.path.append('/content/nltk_data')

# Verify the presence of the resource
#if not os.path.exists(os.path.join(nltk_data_path, 'averaged_perceptron_tagger')):
#    print("Resource not found. Please ensure the file is correctly uploaded.")
#else:
#    print("Resource is correctly uploaded.")

pdf_folder_path = "/content/"
pdf_file_name = "Oral_defense_demo.pdf"


# Extract images, tables, and chunk text from a PDF file.
raw_pdf_elements = partition_pdf(
    filename=pdf_file_name,
    extract_images_in_pdf=False,
    infer_table_structure=True,
    chunking_strategy="by_title",
    max_characters=40000,
    new_after_n_chars=38000,
    combine_text_under_n_chars=2000,
    image_output_dir_path=pdf_folder_path,
)

# Categorize extracted elements from a PDF into tables and texts.
tables = []
texts = []
for element in raw_pdf_elements:
    if "unstructured.documents.elements.Table" in str(type(element)):
        tables.append(str(element))
    elif "unstructured.documents.elements.CompositeElement" in str(type(element)):
        texts.append(str(element))

# Optional: Enforce a specific token size for texts
text_splitter = CharacterTextSplitter.from_tiktoken_encoder(
    chunk_size=10000, chunk_overlap=0
)
joined_texts = " ".join(texts)
texts_4k_token = text_splitter.split_text(joined_texts)
```

```python
#texts_4k_token
# Define the file name
# Split the file name into root and extension
import os.path
file_root, file_extension = os.path.splitext(pdf_file_name)
file_name = file_root+".txt"

# Write the string to the file
with open(file_name, "w") as file:
    file.write(texts_4k_token[0])
```

## 3. Set LLM model and Create Vector Index

```python
#import os.path
from llama_index.core import (
    Settings,
    VectorStoreIndex,
    SimpleDirectoryReader,
    StorageContext,
    load_index_from_storage,
)
from llama_index.llms.gemini import Gemini
from llama_index.embeddings.gemini import GeminiEmbedding
import os
from google.colab import userdata
GEMINI_API = userdata.get('GEMINI_API')

# 更改全域配置
Settings.embed_model = GeminiEmbedding(model_name="models/embedding-001", api_key=GEMINI_API)
Settings.llm = Gemini(api_key=GEMINI_API,model_name="models/gemini-1.5-flash",temperature=0,max_tokens=2048) #top_p=0.85
#https://generativeai.pub/gemini-1-5-flash-in-action-insights-from-a-hands-on-experiment-f46fc5be4deb

#Check and Create Stored Directory
PERSIST_DIR = "./storage"
if not os.path.exists(PERSIST_DIR):
    #reader = SimpleDirectoryReader(input_dir=pdf_folder_path)
    reader = SimpleDirectoryReader(input_files=[pdf_folder_path+file_name])
    documents = reader.load_data()
    #documents = SimpleDirectoryReader("docs").load_data()
    index = VectorStoreIndex.from_documents(documents)
    index.storage_context.persist(persist_dir=PERSIST_DIR)
else:
    storage_context = StorageContext.from_defaults(persist_dir=PERSIST_DIR)
    index = load_index_from_storage(storage_context)
```

```python
#Test the query engine
query_engine = index.as_query_engine()
query = "What is Scorpiahc 1.0?"
response = query_engine.query(query)
print(response)
```
